14. How can you find correlation between variables in Python?
- You can easily find correlation between variables in Python using pandas and Seaborn/Matplotlib for visualization. Here‚Äôs a step-by-step guide:
1. Using pandas.DataFrame.corr()

The corr() function computes the correlation matrix for all numeric columns in a DataFrame.

**Example:**

```python
import pandas as pd

data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 65000, 70000, 75000],
    'Hours_Worked': [40, 42, 38, 45, 50]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()
print(corr_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

15. What is causation? Explain difference between correlation and causation with an example.
     - Causation (or cause-and-effect) means that one variable directly affects another.

        If A causes B, changing A will produce a change in B.

        Causation implies a direct relationship, not just a statistical association.
 | Aspect             | Correlation                                                         | Causation                                           |
| ------------------ | ------------------------------------------------------------------- | --------------------------------------------------- |
| **Definition**     | Measures whether two variables move together (positive or negative) | One variable directly influences the other          |
| **Direction**      | No implication of direction (A ‚Üî B)                                 | Clear direction (A ‚Üí B)                             |
| **Requirement**    | Only a statistical relationship                                     | Requires evidence or mechanism showing cause-effect |
| **Interpretation** | ‚ÄúVariables are related‚Äù                                             | ‚ÄúVariable A causes Variable B‚Äù                      |
| **Example**        | Ice cream sales ‚Üë and drowning ‚Üë in summer                          | Smoking ‚Üí lung cancer                               |


 **Example:-**

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
    - In machine learning and deep learning, an optimizer is an algorithm that adjusts the model‚Äôs parameters (weights and biases) to minimize the loss function during training.

      Goal: Find the set of parameters that produces the lowest possible loss, improving model accuracy or performance.

      Works by calculating the gradient of the loss with respect to parameters and updating them step by step.
      Different Types of Optimizers
    1. Gradient Descent (GD)

     Basic optimizer that updates weights using the gradient of the loss.

     Variants:

     Batch Gradient Descent: Uses the entire dataset to compute gradient in each step.

      Accurate but slow for large datasets.

      Stochastic Gradient Descent (SGD): Uses one sample at a time.

      Faster but noisy, can escape local minima.

      Mini-batch Gradient Descent: Uses a small batch of samples.

       Balances speed and stability.
     
     2. Momentum

     Accelerates SGD by adding a momentum term to the weight update.

     Helps overcome local minima and smooth updates
     Example: Deep learning models in TensorFlow or PyTorch often use SGD with momentum.

      3. Adagrad (Adaptive Gradient)

     Adjusts learning rate for each parameter individually.

     Parameters that appear frequently get smaller updates, rare parameters get larger updates.

     Pros: Good for sparse data (e.g., NLP).
     Cons: Learning rate can shrink too much over time.
     
     4. RMSprop

     Similar to Adagrad but avoids learning rate decay problem by using a moving average of squared gradients.

     Popular in recurrent neural networks (RNNs).

     5. Adam (Adaptive Moment Estimation)

      Combines Momentum + RMSprop.

     Maintains moving averages of gradients and squared gradients.

      Most widely used optimizer today

18. What does model.fit() do? What arguments must be given?
    - In Scikit-Learn, model.fit() is the method used to train a machine learning model.

       It finds the best parameters (weights, coefficients, etc.) for the model using the training data.
       During this process, the model learns the relationship between input features and the target variable.
       
       Work:-  
      
      You pass training data to the model.

      The model computes predictions based on current parameters.

      It calculates the loss/error between predictions and actual target values.

      The model updates its internal parameters to minimize the loss.

      Repeats until convergence (or a maximum number of iterations).
      
      **Arguments of model.fit()**
      1. X (features/input data)

          Array-like (NumPy array, pandas DataFrame) of shape (n_samples, n_features).
          
          Each row = one sample, each column = one feature.

      2. y (target/output data)

         Array-like of shape (n_samples,) for regression or binary classification.

         For multiclass classification, shape = (n_samples,) or (n_samples, n_outputs).

       3. sample_weight (optional)

          Array-like of weights for each sample.

          Useful if some samples should influence the model more than others.

19. What does model.predict() do? What arguments must be given?
    - In Scikit-Learn, model.predict() is used to make predictions using a trained machine learning model.

       After the model has been trained with model.fit(), it applies the learned parameters to new data.

       Returns the predicted outputs (targets) for the given input features.

       **Arguments of model.predict()**
       1. X (input features)

        Array-like (NumPy array, pandas DataFrame, etc.) of shape (n_samples, n_features).

         New data or test data you want predictions for.

       Each row = one sample, each column = one feature.

      Important:
         The number of features in X must match the number of features used during training (model.fit()).

       2. No target (y) required

        Unlike fit(), predict() only needs the input features, not the actual outputs.

        The goal is to generate predicted outputs.

21. What is feature scaling? How does it help in Machine Learning?
     - Feature scaling is the process of standardizing or normalizing the range of independent variables (features) in your dataset so that all features have similar scales.
   
     **Importance of Feature Scaling in Machine Learning**
     1. Speeds up convergence

     Gradient-based algorithms (like linear regression, logistic regression, neural networks) converge faster when features are on a similar scale.

     2. Avoids bias

       Features with larger magnitudes can disproportionately influence models like:

      K-Nearest Neighbors (KNN)

     Support Vector Machines (SVM)

     Principal Component Analysis (PCA)

      3. Improves performance

      Models may achieve better accuracy and more stable results.

24. How do we split data for model fitting (training and testing) in Python?
    - Using train_test_split

     train_test_split is a function from sklearn.model_selection that splits your dataset into training and testing sets.

     **Example:-**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Sample dataset
data = {'Feature1': [10, 20, 30, 40, 50],
        'Feature2': [5, 10, 15, 20, 25],
        'Target': [0, 1, 0, 1, 0]}
df = pd.DataFrame(data)

X = df[['Feature1', 'Feature2']]  # Features
y = df['Target']                  # Target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

print("Training features:\n", X_train)
print("Testing features:\n", X_test)
```

25. Explain data encoding?
    - Data encoding is the process of converting categorical data into numerical format so that machine learning models can process it.

        Most ML algorithms cannot work with text or categorical labels directly.

       Encoding transforms categories into numbers without losing information.
       Types of Data Encoding
       1. Label Encoding

       Assigns a unique integer to each category.

        Useful for ordinal categories (where order matters).
        
        2. One-Hot Encoding

       Creates binary columns for each category.

        Avoids implying any order.

        Useful for nominal categorical variables.
        
        3. Ordinal Encoding

        Similar to label encoding but respects the order of categories.

        Example: Education Level ‚Üí {High School: 0, Bachelor: 1, Master: 2, PhD: 3}

       4. Binary / Dummy Encoding

       Converts categorical variable into binary numbers.

        Often used for high-cardinality features to reduce dimensionality.
