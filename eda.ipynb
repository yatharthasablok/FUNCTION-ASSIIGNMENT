```python
q1= df.km_driven.quantile(.25)
q3= df.km_driven.quantile(.75)
IQR = q3-q1
lower_bound =  q1 - 1.5* IQR
upper_bound =  q3 + 1.5* IQR

outliers_df =  df[(df.km_driven<lower_bound) | (df.km_driven>upper_bound)]
outliers_df
```

```python
df_remove =  df[(df.km_driven>=lower_bound) & (df.km_driven<=upper_bound)]
print(f"df shape before outlier removal is {df.shape}")
print(f"df shape after outlier removal is {df_remove.shape}")
```

6. What are the most popular product keywordsk


```python
from collections import Counter
import re


all_product_names = " ".join(df["product_name"].str.lower())

words = re.findall(r'\b[a-z]+\b', all_product_names)

word_counts = Counter(words)

top_keywords = word_counts.most_common(20)
print(top_keywords)
```

```python
keywords, counts = zip(*top_keywords)
plt.figure(figsize=(12,6))
plt.bar(keywords, counts, color='skyblue')
plt.xticks(rotation=45)
plt.title("Top 20 Product Keywords")
plt.ylabel("Frequency")
plt.show()
```
