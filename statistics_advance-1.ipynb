### **Question 1 — What is a random variable in probability theory?**
A **random variable** is a function that assigns a numerical value to each outcome of a random experiment. It maps outcomes in the sample space to real numbers.

---

### **Question 2 — What are the types of random variables?**
1. **Discrete random variables** — take countable values (e.g., 0,1,2,...).  
2. **Continuous random variables** — take values from an interval (uncountably many), described by a density function.  
3. **Mixed random variables** — have both discrete and continuous components.

---

### **Question 3 — Explain the difference between discrete and continuous distributions.**
- **Discrete distribution:** probability mass function (PMF) gives \(P(X=x)\); probabilities sum to 1. Example: Binomial.  
- **Continuous distribution:** probability density function (PDF) \(f(x)\) satisfies \(P(a \le X \le b)=\int_a^b f(x)dx\); \(P(X=x)=0\). Example: Normal.

---

### **Question 4 — What is a binomial distribution, and how is it used?**
**Binomial distribution** models number of successes in \(n\) independent Bernoulli trials each with success probability \(p\):  
\[ P(X=k)=\binom{n}{k} p^k (1-p)^{n-k} \]  
Used for yes/no outcomes repeated a fixed number of times.

---

### **Question 5 — What is the standard normal distribution, and why is it important?**
**Standard normal:** normal with mean 0 and sd 1, denoted \(Z \sim N(0,1)\).  
It’s used for z-scores; any \(X \sim N(\mu, \sigma^2)\) converts via \(Z=(X-\mu)/\sigma\).

---

### **Question 6 — What is the Central Limit Theorem (CLT), and why is it critical?**
**CLT:** For large \(n\), the sampling distribution of \(\bar{X}\) is approximately normal regardless of population distribution (finite variance).  
\( \bar{X} \sim N(\mu, \sigma^2/n) \) approximately.  
**Importance:** Enables normal-based inference even for non-normal populations.

---

### **Question 7 — Significance of confidence intervals**
A **confidence interval (CI)** gives a range of plausible values for a parameter.  
A 95% CI means 95% of such intervals from repeated samples contain the true mean.  
It quantifies estimation uncertainty.

---

### **Question 8 — Concept of expected value**
Expected value \(E[X]\) (mean) is the long-run average of outcomes.  
- Discrete: \(E[X]=\sum_x x P(X=x)\)  
- Continuous: \(E[X]=\int x f(x)dx\)


### **Question 9 — Python Program for Normal Distribution**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 normal random numbers with mean=50 and sd=5
np.random.seed(0)
q9_data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviations
q9_mean = q9_data.mean()
q9_std_pop = q9_data.std(ddof=0)
q9_std_sample = q9_data.std(ddof=1)

print(f"Mean = {q9_mean:.6f}")
print(f"Population Std (ddof=0) = {q9_std_pop:.6f}")
print(f"Sample Std (ddof=1) = {q9_std_sample:.6f}")

plt.figure(figsize=(8, 4))
plt.hist(q9_data, bins=30 , color="g")
plt.title("Q9: Histogram of 1000 Samples from Normal(50, 5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
````

### **Question 10 — Daily sales and CLT-based 95% Confidence Interval**
Given daily sales data:
```python
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
```
We calculate mean, standard deviation, standard error, and 95% CI using z=1.96.


````python
import numpy as np, math


daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

n = len(daily_sales)
mean_sales = np.mean(daily_sales)
s_sample = np.std(daily_sales, ddof=1)
z_95 = 1.96

se = s_sample / math.sqrt(n)
ci_low = mean_sales - z_95 * se
ci_high = mean_sales + z_95 * se

print(f"n = {n}")
print(f"Mean sales = {mean_sales:.6f}")
print(f"Sample standard deviation (s) = {s_sample:.6f}")
print(f"Standard error (s/√n) = {se:.6f}")
print(f"95% CI for mean (using z=1.96) = [{ci_low:.6f}, {ci_high:.6f}]")
````
