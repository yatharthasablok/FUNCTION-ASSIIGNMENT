## Question 1
Hypothesis testing is a method to determine whether sample data provides enough evidence to reject a claim about a population.

## Question 2
The null hypothesis (H₀) states no effect or no difference.  
The alternative hypothesis (H₁) states an effect or difference exists.

## Question 3
The significance level (α) is the probability of rejecting a true null hypothesis. If p-value ≤ α, we reject H₀.

## Question 4
Type I error: Rejecting a true H₀.  
Type II error: Not rejecting a false H₀.

## Question 5
Z-test is used when population standard deviation is known or sample size is large.  
T-test is used when population standard deviation is unknown and sample size is small.

## Question 7

```python

import numpy as np
from scipy.stats import norm

data = np.array([49.1,50.2,51.0,48.7,50.5,49.8,50.3,50.7,50.2,49.6,
50.1,49.9,50.8,50.4,48.9,50.6,50.0,49.7,50.2,49.5,
50.1,50.3,50.4,50.5,50.0,50.7,49.3,49.8,50.2,50.9,
50.3,50.4,50.0,49.7,50.5,49.9])

mean = data.mean()
std = data.std(ddof=1)
z = (mean - 50) / (std / np.sqrt(len(data)))
p = 2 * (1 - norm.cdf(abs(z)))
z, p
```

## Question 8

```python

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)
mean = data.mean()
std = data.std(ddof=1)
ci_low = mean - 1.96 * std / np.sqrt(len(data))
ci_high = mean + 1.96 * std / np.sqrt(len(data))

plt.hist(data)
plt.show()

(ci_low, ci_high)
```
