Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer:

Descriptive Statistics summarize and describe the features of a dataset. They focus on what the data shows.
Examples: Mean, Median, Mode, Standard Deviation, Graphs (like histograms, boxplots).

Example:
If you calculate the average test score of a class of 40 students, that’s descriptive statistics.

Inferential Statistics make predictions or inferences about a population based on a sample.
They use probability theory to estimate or test hypotheses.

Example:
If you survey 100 voters and predict the election results for the entire country, that’s inferential statistics.

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer:

Sampling is the process of selecting a subset (sample) from a larger group (population) to draw conclusions about the entire population.

Random Sampling:
Each member of the population has an equal chance of being selected.
✅ Example: Choosing 100 names randomly from a database of 10,000 customers.

Stratified Sampling:
The population is divided into groups (strata) based on characteristics (e.g., age, gender), and random samples are taken from each group proportionally.
✅ Example: Dividing a population into “male” and “female” groups and randomly selecting 50 from each.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:

Mean: The arithmetic average of all values.
Formula: sum of all values divied by total number of values

	​


Median: The middle value when all observations are arranged in order.

Mode: The value that occurs most frequently.

Importance:
These measures represent the central tendency of a dataset — they give an idea about the “typical” value and help in comparing data distributions.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer:

Skewness measures the asymmetry of the data distribution.

Positive Skew: Tail is longer on the right side → mean > median.

Negative Skew: Tail is longer on the left side → mean < median.

Kurtosis measures the peakedness or flatness of a distribution.

High kurtosis: Sharp peak (more outliers)

Low kurtosis: Flatter distribution

✅ Positive Skew means most data points are low values, but a few large values pull the tail to the right.

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets.
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

cov_matrix = np.cov(list_x, list_y)
covariance = cov_matrix[0, 1]
correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
"""Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]"""
import matplotlib.pyplot as plt
import numpy as np


survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

mean = np.mean(survey_scores)
std = np.std(survey_scores)

plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()

print("Mean:", mean)
print("Standard Deviation:", std)
```
